### (a) Plot these six training points. Are the classes {+, −} linearly separable? 

```python
data = np.array([[1, 1],
                 [2, 2],
                 [2, 0]])

x1, y1 = data.T

plt.scatter(x1, y1, color = 'red')

for x2, y2 in zip(x1,y1):
    label = f"({x2},{y2})(+)"
    plt.annotate(label, (x2, y2), textcoords="offset points", xytext=(0,10), ha='center')


data = np.array([[0, 0],
                 [1, 0],
                 [0, 1]])
x1, y1 = data.T

plt.scatter(x1, y1, color = 'blue')

for x2, y2 in zip(x1, y1):
    label = f"({x2}, {y2})(-)"
    plt.annotate(label, (x2, y2), textcoords="offset points", xytext=(0,10), ha='center')

x1 = np.linspace(-1, 3, 100)
y1 = x1*0

plt.plot(x1, y1, 'grey', linewidth=0.5)

y1 = np.linspace(-1, 3, 100)
x1 = y1*0
plt.plot(x1, y1, 'grey', linewidth=0.5)



x1 = np.linspace(-1, 3, 100)
y1 = x1 * -1 + 1

plt.plot(x1, y1, 'blue', linewidth=0.5)

x1 = np.linspace(-1, 3, 100)
y1 = x1 * -1 + 2

plt.plot(x1, y1, 'red', linewidth=0.5)

x1 = np.linspace(-1, 3, 100)
y1 = x1 * -1 + 1.5

plt.plot(x1, y1, 'black')

plt.savefig("task5a.png", dpi=300)

plt.show()
```

We can see in the above graph that the classes {+, −} are linearly separable. The classes are separable by the maximum hyperplane formed by the line $x_1 + x_2 = 3/2$

### (b) Consider mapping each point to 3-D using new feature vectors $\phi(x) = [1, \sqrt{2}.x, x^2]$. Are the classes now linearly separable? If so, find a separating hyperplane.

New feature vectors, based on $\phi(x) = [1, \sqrt{2}.x, x^2]$, are: $(1, 0, 0), (1, − \sqrt{2}, 1), (1, \sqrt{2}, 1)$ 

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

root2 = sqrt(2)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')


for xs, ys, zs in [(1, root2, 1), (1, -1 * root2, 1)]:
    ax.scatter(xs, ys, zs, color='blue')

for xs, ys, zs in [(1, 0, 0)]:
    ax.scatter(xs, ys, zs, color='red')

plt.savefig("task6b.png", dpi=300)

plt.show()
```

From the above graph, we can see that (-) negatives are in $z=1$ plane and (+) positive point is $z=0$ plane. The classes are now linearly separable in 3-dimensional space. 

A separating hyperplane is given by the weight vector (0, 0, 1) in the new space. 

## Task 7: Learning SVMs on the Titanic dataset. Please report your five-fold cross validation classification accuracies on Titanic training set, with respect to the linear, quadratic, and RBF kernels. Which kernel is the best in your case? 

```python
# load the Titanic dataset

train_df = pd.read_csv("Data/train.csv", header=0)
test_df = pd.read_csv('Data/test.csv', header=0)

combine_df = [train_df, test_df]
```

### Preprocess the Titanic dataset

```python
# check the missing values
def missing_percentage(df):
    total = df.isnull().sum().sort_values(ascending = False)
    percent = round(df.isnull().sum().sort_values(ascending = False)/len(df)*100, 2)
    return pd.concat([total, percent], axis=1, keys=['Total','Percent'])

missing_percentage(train_df)
```

```python
# dropping the columns which are not useful
train_df = train_df.drop(['Ticket', 'Cabin'], axis=1)
test_df = test_df.drop(['Ticket', 'Cabin'], axis=1)
combine_df = [train_df, test_df]
```

```python
# Creating new feature from existing
for data in combine_df:
    data['Title'] = data.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
# replace many titles with a more common name or classify them as Other
for data in combine_df:
    data['Title'] = data['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Other')

    data['Title'] = data['Title'].replace('Mlle', 'Miss')
    data['Title'] = data['Title'].replace('Ms', 'Miss')
    data['Title'] = data['Title'].replace('Mme', 'Mrs')
    
train_df[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
# convert the categorical titles to ordinal
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Other": 5}

for data in combine_df:
    data['Title'] = data['Title'].map(title_mapping)
    data['Title'] = data['Title'].fillna(0)

train_df.head()
```

```python
# Create a new feature for FamilySize that combines Parch and SibSp features and drop Parch and SibSp from dataset
for data in combine_df:
    data['FamilySize'] = data['SibSp'] + data['Parch'] + 1

train_df[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)


train_df = train_df.drop(['Name', 'PassengerId'], axis=1)
test_df = test_df.drop(['Name'], axis=1)

combine_df = [train_df, test_df]
```

```python
train_df = train_df.drop(['Parch', 'SibSp'], axis=1)
test_df = test_df.drop(['Parch', 'SibSp'], axis=1)
combine_df = [train_df, test_df]
```

```python
# change the sex data from categorical to numerical
for data in combine_df:
    data['Sex'] = data['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
```

```python
# filling missing values in the Age and Fare columns using the median values of the data
train_df['Fare'] = train_df['Fare'].fillna(train_df['Fare'].dropna().median())
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].dropna().median())

# filling the missing values in the Embarked columnn with the most common embarked port 'S'
port_freq = train_df['Embarked'].dropna().mode()[0]
train_df['Embarked']= train_df['Embarked'].fillna(port_freq)

train_df[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
# convert the embarked data from categorical to numerical
for data in combine_df:
    data['Embarked'] = data['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)

train_df.head()
```

```python
# select important features based on HW2
selected_features = ['Pclass','Sex','Fare', 'Title','Embarked','Age', 'FamilySize']
X = train_df[selected_features]
y = train_df['Survived']
```

```python
# split the data into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

### Please report your five-fold cross validation classification accuracies on Titanic training set, with respect to the linear, quadratic, and RBF kernels. Which kernel is the best in your case?

```python
# run five-fold cross validation test on Titanic data using SVM kernels (linear, rbf and quadratic)

clf = SVC(kernel='linear')
lin_acc_scores = cross_val_score(clf, X=X_train, y=y_train, cv=5) 
print('Linear kernel    ', lin_acc_scores, ", Average = ", lin_acc_scores.mean())

clf = SVC(kernel='rbf')
rbf_acc_scores = cross_val_score(clf, X=X_train, y=y_train, cv=5)
print('RBF kernel       ', rbf_acc_scores, ", Average = ", rbf_acc_scores.mean())

clf = SVC(kernel='poly', degree=2)
poly_acc_scores = cross_val_score(clf, X=X_train, y=y_train, cv=5)
print('Quadratic kernel ', poly_acc_scores, ", Average = ", poly_acc_scores.mean())
```

```python
scores_total = []

for kernel in ('linear', 'poly', 'rbf'):
    if kernel == 'poly':
        clf = SVC(kernel=kernel, degree=2)
    else:
        clf = SVC(kernel=kernel)
    scores_total.append(cross_val_score(estimator=clf, X=X_train, y=y_train, cv=5))
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(kernel + ' classification report:')
    print(classification_report(y_test, y_pred))
    
# SVM kernels cross-validation scores comparison
fig = plt.figure()
fig.suptitle('Five-fold cross validation scores comparison for SVM kernels')
ax = fig.add_subplot(111)
plt.boxplot(scores_total)
ax.set_xticklabels(['Linear', 'Polynomial (Quadratic)', 'RBF'])
plt.savefig("SVM-kernals-comparison-without-tuning.png", dpi=300)
plt.show()
```

### Tuning the parameters and applying `GridSearchCV` for each SVM kernels

```python
# linear kernel

svm_params = {'kernel': ['linear'], 'C': [0.1, 1, 10, 50, 100], 'probability': [True]}

svm = GridSearchCV(SVC(), svm_params, cv=5)
svm.fit(X_train, y_train)

print("Best accuracy: ", svm.best_score_)
print("Best parameters: ", svm.best_params_)
```

```python
# RBF kernel

svm_params = {'kernel': ['rbf'], 'C': [1, 10, 50, 100, 200], 'gamma': [1, 0.1, 0.01, 0.001], 'probability': [True]}

svm = GridSearchCV(SVC(), svm_params, cv=5)
svm.fit(X_train, y_train)

print("Best accuracy: ", svm.best_score_)
print("Best parameters: ", svm.best_params_)
```

```python
# polynomial (quadratic) kernel

svm_params = {'kernel': ['poly'], 'C': [1, 10, 50, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'degree': [2], 'probability': [True]}

svm = GridSearchCV(SVC(), svm_params, cv=5)
svm.fit(X_train, y_train)

print("Best accuracy: ", svm.best_score_)
print("Best parameters: ", svm.best_params_)
```
